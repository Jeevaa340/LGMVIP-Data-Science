### Task 4 -Image to Pencil Sketch with Python

We need to read the image in RBG format and then convert it to a grayscale image. This will turn an image into a classic black and white photo. Then the next thing to do is invert the grayscale image also called negative image, this will be our inverted grayscale image. Inversion can be used to enhance details. Then we can finally create the pencil sketch by mixing the grayscale image with the inverted blurry image. This can be done by dividing the grayscale image by the inverted blurry image. Since images are just arrays, we can easily do this programmatically using the divide function from the cv2 library in Python.

## Importing Packages

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn')
```

## Loading Images

```python
img = cv2.imread(r'image 1.jpg')
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.figure(figsize=(8,8))
plt.imshow(img)
plt.axis("off")
plt.title("Original Image")
plt.show()
```

## Resizing image shape

```python
scale_percent = 0.60
width = int(img.shape[1]*scale_percent)
height = int(img.shape[0]*scale_percent)
```

```python
dim = (width,height)
resized = cv2.resize(img,dim,interpolation = cv2.INTER_AREA)

plt.figure(figsize=(8,8))
plt.imshow(resized)
plt.axis("off")
plt.title("Resized Image")
plt.show()
```

## Converting an image into gray_scale image

```python
gray = cv2.cvtColor(resized , cv2.COLOR_BGR2GRAY)
plt.figure(figsize=(8,8))
plt.imshow(gray,cmap='gray')
plt.axis("off")
plt.title("gray Image")
plt.show()
```

## Inverting the image

```python
inv = 255-gray
plt.figure(figsize=(8,8))
plt.imshow(inv,cmap='gray')
plt.axis("off")
plt.title("invy Image")
plt.show()
```

## Smoothing the image

```python
gauss = cv2.GaussianBlur(inv,ksize=(15,15),sigmaX=0,sigmaY=0)
plt.figure(figsize=(8,8))
plt.imshow(gauss,cmap='gray')
plt.axis("off")
plt.title("gauss Image")
plt.show()
```

## Obtaining the final sketch

```python
def dodgeV2(image,mask):
    return cv2.divide(image,255-mask,scale=256)
```

```python
pencil_img = dodgeV2(gray,gauss)

plt.figure(figsize=(8,8))
plt.imshow(pencil_img,cmap='gray')
plt.axis("off")
plt.title("invy Image")
plt.show()
```

## comparing images of all steps

```python
plt.figure(figsize=(20,20))
plt.subplot(1,7,1)
plt.imshow(img)
plt.axis("off")
plt.title("Original Image")
plt.subplot(1,7,2)
plt.imshow(resized)
plt.axis("off")
plt.title("Resized Image")
plt.subplot(1,7,3)
plt.imshow(gray,cmap="gray")
plt.axis("off")
plt.title("GrayScale Image")
plt.subplot(1,7,4)
plt.imshow(inv,cmap="gray")
plt.axis("off")
plt.title("Inverted Image")
plt.subplot(1,7,5)
plt.imshow(gauss,cmap="gray")
plt.axis("off")
plt.title("Smoothen Image")
plt.subplot(1,7,6)
plt.imshow(pencil_img,cmap="gray")
plt.axis("off")
plt.title("Final Sketch Image")
plt.show()
```
